```python
up_date_data = all_df.copy()
up_date_data.head()
```

```python
up_date_data.tail()
```

```python
def myfunc(acc, title):
    if pd.notnull(acc):
        acc = acc
    elif (title == 'Sales Assistant') and pd.isnull(acc):
        acc = 'Contract Entry'
    elif (title == 'Account Manager') and pd.isnull(acc):
        acc = 'Account Admin'
    elif (title == 'Director of Sales') and pd.isnull(acc):
        acc = 'Approval - Manager'
    return acc
```

```python
up_date_data['NewColumn_1'] = up_date_data.apply(lambda x: myfunc(x['access_level'], x['title']), axis=1)
up_date_data.head()
```

```python
up_date_data.tail()
```

```python
up_date_data.to_csv('up_date_data.csv', index=False) #data updated with appropriate access for confirmation
```

```python
#reorder and rename
master_data = up_date_data[['id','NewColumn_1','first_name','last_name','email','title']]
master_data.columns = ['id','access_level','first_name','last_name','email','title']
master_data.head()
```

```python
#Write to File
master_data.to_csv('MasterData.csv', index=False) #All data updated with appropriate columns
```

```python
## Missing Data
miss_data = missing_data.copy()
miss_data['access_level'] = miss_data.apply(lambda x: myfunc(x['access_level'], x['title']), axis=1)
miss_data
```

```python
#Write to File
missing_data.to_csv('missingData.csv', index=False) #Missing data that neeeds updating
#Write to File
miss_data.to_csv('missWithUpdateData.csv', index=False) #Missing data with Updates
```
